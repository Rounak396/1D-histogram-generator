```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cancer patient data sets.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['n_level'] = le.fit_transform(df.Level)
```

```python
df.drop(['Patient Id'],axis='columns',inplace=True)
```

```python
df.sample(6)
```

```python
df.drop(index=df[df['Age'] > 68].index, inplace=True)
```

```python
df.drop(['Age', 'Swallowing Difficulty'], axis='columns',inplace=True)
```

```python
sns.heatmap (df.corr())
```

```python
df.shape
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
```

```python
df.drop(['Level'],axis='columns', inplace=True)
```

```python
X = df.drop(['n_level'],axis='columns')
```

```python
Y = df['n_level']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state=10,test_size=0.2)
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression(max_iter= 1000, tol= 0.001)
```

```python
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier(n_estimators=50)
```

```python
model3.fit(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```
